```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
X_train_prediction=model.predict(X_train)
```

```python
X_train_accuracyscore=accuracy_score(X_train_prediction,Y_train)
```

```python
print("Accuracy of train data set = ",X_train_accuracyscore)
```

```python
X_test_prediction=model.predict(X_test)
```

```python
X_test_accuracyscore=accuracy_score(X_test_prediction,Y_test)
```

```python
print("Accuracy score of the test data set = ",X_test_accuracyscore)
```
